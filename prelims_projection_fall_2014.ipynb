### Project a social network from the Prelims graph using ProjX (https://pypi.python.org/pypi/ProjX)

```python
# Again we'll use the class projx.Projection
p = px.Projection(g, node_type_attr='[Schema] Type')
```

#### STEPS 1 - 2 are basically clean up. We will perform them over the full graph until we reach the final projection.

#### STEP 1: First we want to conserve the city of publication by transferring the attributes of that entity onto the the edition, which will form the heart of the projection. This is how we perserve a lot of the richness from the original model.

```python
proj = p.execute('TRANSFER ATTRS (l:Lugar)-(i:Impresion)-(e:Edicion)')
```

```python
# Here is what we get
print(proj.node['1939'])
```

#### STEP 2: We need to project all individuals that collaborated on a edition through their respective documents to create a direct connection to the edition. We'll also transfer document attributes to the edition node to preserve this info for future processing. 

```python
p1 = px.Projection(proj, node_type_attr='[Schema] Type')  # Maybe I should return an instance of projection in the future?
proj1 = p1.execute('''
    PROJECT (p:Persona)-(wild)-(e:Edicion)
    TRANSFER ATTRS (wild)-(e)
''')
```

```python
nodes = proj1.neighbors('1939')
print([g.node[node]['label'] for node in nodes])
```

<h4>Now:  
1. City attributes have been transfered to editions.  
2. Front matter attributes have been transfered to editions.    
3. Anyone who participated in the preliminaries is connected to the edition.  
Just two last steps...</h4>

#### STEP 3: Match a subgraph that contains just the editions and the people connected to them. Then we project accross the editions. The library calculates edge weight as a Jaccard Index by default on projection.

```python
p2 = px.Projection(proj1, node_type_attr='[Schema] Type')
proj2 = p2.execute('''
    MATCH (p1:Persona)-(e:Edicion)-(p2:Persona)
    PROJECT (p1)-(e)-(p2)
''')
```

#### STEP 4: Finally, we will remove the low weight edges using a cutoff that reduces overall graph density.

```python
cutoffs, densities = px.proj_density(proj2, 0.02, 0.02, 49)
plt.plot(cutoffs, densities)
```

```python
proj3 = px.remove_edges(proj2.copy(), 0.2)
print(len(proj3), len(proj3.edges()))
```

#### The ingenioso social circle of Cervantes.

```python
# Again, just a couple quick manipulations.
p3 = px.Projection(proj3, node_type_attr='[Schema] Type')
nbrs = proj3.neighbors('1930')
edges = zip(['1930'] * len(nbrs), nbrs)
graph = p3.build_subgraph(edges)
px.draw_simple_graph(graph, node_type_attr='[Schema] Type')
```
